```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\rizky\Semester_3\Data_Science\calon_tni.csv")

print('=== Data Calon TNI ===')
print(df)

print('\n=== Statistik Dasar Tinggi Badan ===')
print(df['tinggi'].describe())

print('Rata-rata tinggi : 175.70 cm')
print('Tertinggi : 185 cm')
print('Terendah : 160 cm')
print('Median: 178.0 cm')

print('\n=== Calon di atas rata-rata tinggi ===')
df_mean = df[df['tinggi'] > 175.70]
print(df_mean)

print('\n=== Urutan dari Tertinggi ke Terendah ===')
df_sort = df.sort_values(by='tinggi', ascending=False)
print(df_sort)

print('\n=== Data dengan Kategori ===')
def kategori_tinggi(x):
    if x >= 180:
        return 'Tinggi'
    elif x >= 170:
        return 'Sedang'
    else:
        return 'Pendek'

df['kategori'] = df['tinggi'].apply(kategori_tinggi)
print(df)

print("\n=== Jumlah Calon per Kategori ===")
print(df['kategori'].value_counts())

print("\n=== Statistik Lanjutan ===")
print("Standar Deviasi :", round(df['tinggi'].std(), 2))
print("Variansi :", round(df['tinggi'].var(), 2))

print("\n=== Calon dengan Tinggi Badan Tertinggi ===")
tertinggi = df.loc[df['tinggi'].idxmax()]
print("Nama   :", tertinggi['nama'])
print("Alamat :", tertinggi['alamat'])
print("Tinggi :", tertinggi['tinggi'], "cm")
```

```python
print("\n=== Deteksi Masalah Data ===")
print(df.isnull().sum())
print("Jumlah duplikat:", df.duplicated().sum())
```
